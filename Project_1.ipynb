They have supplied you with a csv file called `"bank_marketing.csv"`, which you will need to clean, reformat, and split the data, saving three final csv files. Specifically, the three files should have the names and contents as outlined below:

| column | data type | description | cleaning requirements |

Split and tidy bank_marketing.csv, storing as three DataFrames called client, campaign, and economics, each containing the columns outlined in the notebook and formatted to the data types listed.

Save the three DataFrames to csv files, without an index, as client.csv, campaign.csv, and economics.csv respectively.

## IMPORTING LIBRARIES

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## LOADING THE DATASET


```python
# Load the dataset
Bank_dataset = pd.read_csv("bank_marketing.csv")
```

```python
# Previewing the dataset
Bank_dataset.head()
```

## SUBSETTING DATA FOR EACH REQUIRED FILE

```python
# Subsetting for client.csv
client_columns = ["client_id", "age", "job", "marital", "education", "credit_default", "mortgage"]
```

```python
# Subsetting for campaign.csv
campaign_columns = ["client_id", "number_contacts", "contact_duration", "previous_campaign_contacts", "previous_outcome", "campaign_outcome", "day", "month"]
```

```python
# Subsetting for economics.csv
economics_columns = ["client_id", "cons_price_idx", "euribor_three_months"]
```

```python
# Creating various subsets for the apportioned columns
client = Bank_dataset[client_columns].copy()
campaign = Bank_dataset[campaign_columns].copy()
economics = Bank_dataset[economics_columns].copy()
```

```python
# Previewing the client subset
client.head()
```

```python
# Previewing the campaign subset
campaign.head()
```

```python
# Previewing the economics subset
economics.head()
```

## DATA CLEANING FOR EACH FILE

* DATA CLEANING FOR THE CLIENT DATAFRAME
* Cleaning *client.csv* DataFrame
    * Replace "." with "_" in the job column.
    * Change "unknown" values in education to NaN.
    * Convert *credit_default* and mortgage columns to boolean types.


```python
# Clean client DataFrame
client['job'] = client['job'].str.replace(".", "_", regex=False)
client['education'] = client['education'].replace("unknown", np.nan)
client['credit_default'] = client['credit_default'].astype(bool)
client['mortgage'] = client['mortgage'].astype(bool)
```

```python
client.head()
```

* DATA CLEANING FOR CAMPAIGN DATAFRAME
* Cleaning *campaign.csv* DataFrame
    * Convert previous_outcome and campaign_outcome columns to boolean types.
    * Create a last_contact_date column by combining day, month, and setting the year as 2022.

```python
# Clean campaign DataFrame
campaign['previous_outcome'] = campaign['previous_outcome'].astype(bool)
campaign['campaign_outcome'] = campaign['campaign_outcome'].astype(bool)

# Create the last_contact_date column
campaign['last_contact_date'] = pd.to_datetime(
    '2022' + campaign['month'] + campaign['day'].astype(str),
    format='%Y%b%d'
)
# Drop the separate day and month columns
campaign.drop(['day', 'month'], axis=1, inplace=True)
```

```python
campaign.head()
```

## SAVE EACH DATAFRAME TO CSV

```python
client.to_csv("client.csv", index=False)
campaign.to_csv("campaign.csv", index=False)
economics.to_csv("economics.csv", index=False)
```
